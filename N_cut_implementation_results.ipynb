### K-way segmentation loss

We can profit from the similarity graph to build a Soft N-cut Loss metric from a k-way (probabilistic) segmentation.

Hard segmentations (pixel labels) can be converted into soft segmentations (K-way) without information loss.

```python
import networkx as nx
from scipy import sparse

def DW_matrix(graph):
    W = nx.to_scipy_sparse_matrix(graph, format='csc')
    entries = W.sum(0)
    D = sparse.dia_matrix( (entries,0),shape = W.shape).tocsc()
    m,n = W.shape
    return D,W

# D,W = DW_matrix(g)  
```

Make a method that computes the N-cut Loss in numpy according to the formula
$$\mathcal{L}(A_1,\ldots,A_K)=K - \sum\limits_{k=1}^K \frac{\sum\limits_{u,v} p(u\in A_k) p(v\in A_k) w(u,v)}{\sum\limits_{u,v} p(u\in A_k) w(u,v)}$$

```python
def ncut_loss(soft_mask ,W , region_labels):
  K=soft_mask.shape[2]
  regions = W.shape[0]
  region_sizes = np.zeros(regions)
  region_soft_max = np.zeros((regions, K))
#   print('average the class probability across pixels of the same region')
  for i in range(soft_mask.shape[0]):
    for j in range(soft_mask.shape[1]):
      region=region_labels[i,j]
      region_soft_max[region]+=soft_mask[i,j]
      region_sizes[region]+=1
  region_soft_max/=region_sizes[:,None]
#   print('compute the loss')
  loss =float(K)
  non_zero = W.nonzero()
  for k in range(K):
    #print('term', k+1, '/', K)
    num = 0
    den=0
    for u,v in zip(non_zero[0],non_zero[1]):
      p_u=region_soft_max[u,k]
      p_v=region_soft_max[v,k]
      num += p_u*p_v*W[u,v]
      den += p_u*W[u,v]
    loss -= num/den
  return loss
```

Convert a image with regions with identical color into  integer labels

```python
def rgb_to_int_labels(rgb_mask):
  int_mask = np.zeros((rgb_mask.shape[0], rgb_mask.shape[1]))
  rgb_labels = {}
  K=0
  for i in range(rgb_mask.shape[0]):
    for j in range(rgb_mask.shape[1]):
      c = rgb_mask[i,j]
      rgb = (c[0],c[1],c[2])
      if rgb not in rgb_labels:
        rgb_labels[rgb]=K
        K+=1
      int_mask[i,j]=rgb_labels[rgb]
#   print(K,'classes')
  return int_mask, K

# ground_segmented_img_int, K = rgb_to_int_labels(ground_segmented_img)
```

Convert a label segmentation into a k-way segmentation.

```python
def mask_to_soft_mask(mask, label_set = None):
  if label_set is None:
    label_set = set([])
    for i in range(mask.shape[0]):
      for j in range(mask.shape[1]):
        #rgb = tuple(mask[i,j])
        label_set.add(mask[i,j])
  
  K = len(label_set)
  label_list = list(label_set)
#   print(K, 'classes')
  soft_mask = np.zeros((mask.shape[0], mask.shape[1], K))
  for k in range(K):
    #print(k)
    bools = mask == label_list[k]
    soft_mask[bools,k] = 1
  return soft_mask

# ground_segmented_img_soft_mask = mask_to_soft_mask(ground_segmented_img_int, label_set = set(range(K)))
```

```python
for img_id in example_ids:
  img =imread('data/original/%s.jpg'%img_id)
  #label each pixel to a region
  labels1 = segmentation.slic(img, compactness=30, n_segments=400)
  out1 = color.label2rgb(labels1, img, kind='avg')
  #build the graph
  g = graph.rag_mean_color(img, labels1, mode='similarity')
#   print(g.number_of_nodes(),'nodes')
  #perform the graph cut
  labels2 = graph.cut_normalized(labels1, g)
  out2 = color.label2rgb(labels2, img, kind='avg')
  
  #profit from the graph to compute loss functions
  D,W = DW_matrix(g)
  
  #get the value of the loss function
  out2_soft = mask_to_soft_mask(labels2)
  n_cut_loss_on_n_cut_segm=ncut_loss(out2_soft, W, region_labels = labels1)
#   print(n_cut_loss_on_n_cut_segm)
  
  #compare with ground truth
  ground_segmented_img=imread('data/segmented_images/%s.png'%img_id)
  ground_segmented_img_int, K = rgb_to_int_labels(ground_segmented_img) 
  ground_segmented_img_soft_mask = mask_to_soft_mask(ground_segmented_img_int, label_set = set(range(K)))  
  n_cut_loss_on_ground_truth=ncut_loss(ground_segmented_img_soft_mask, W, region_labels = labels1)
#   print(n_cut_loss_on_ground_truth)
  
  #plot
  plt.figure(figsize=(20,60))
  
  plt.subplot(1,4,1)
  plt.imshow(img)
  plt.title('original'),plt.xticks([]), plt.yticks([])
  
  plt.subplot(1,4,2)
  plt.imshow(out1)
  plt.title(str(g.number_of_nodes())+' regions'),plt.xticks([]), plt.yticks([])
  
  plt.subplot(1,4,3)
  plt.imshow(out2)
  plt.title('N-cut. Loss '+str(round(n_cut_loss_on_n_cut_segm,3))),plt.xticks([]), plt.yticks([])
  
  plt.subplot(1,4,4)
  plt.imshow(ground_segmented_img)
  plt.title('Ground truth. Loss '+str(round(n_cut_loss_on_ground_truth,3))),plt.xticks([]), plt.yticks([])
  
  plt.show()
```

The evaluation of the N-cut loss on the N-cut segmentation is much smaller than evaluating the N-cut loss on the ground truth, since the N-cut segmentation minimizes the N-cut loss.

This tells us that the N-cut loss does not exactly correspond to the semantic loss.
